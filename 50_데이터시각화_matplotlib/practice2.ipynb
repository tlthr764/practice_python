```python
# 실습2 : 막대그래프 그리기
import matplotlib.pyplot as plt # Matplotlib 불러오기
import numpy as np # numpy 불러오기

#변수
categories = ['Category 1','Category 2','Category 3','Category 4','Category 5']
data = [20, 35, 15, 27, 45]
# 그래프 그리기 : plt.bar([x값],[y값],color,width)
plt.bar(categories,data,color=['r','g','b','y','k'], width=0.6)
# 눈금 표시 : plt.xticks([x값],[눈금]) // plt.yticks([y값],[눈금])
plt.xticks(rotation=45)
# 좌표축 한계 정하기
plt.ylim(0,50)
# 축 레이블
plt.xlabel('Categories',labelpad=30)
plt.ylabel('Values',labelpad=30)
# 그래프 제목
plt.title('Bar Chart')
# 그리드 표시
plt.grid()
# 그래프 출력
plt.show()
```
