```python
# 실습4 : sin & cos 함수 그리기
import matplotlib.pyplot as plt # Matplotlib 불러오기
import numpy as np # numpy 불러오기
import math

# 변수
x = np.linspace(0,10,100)
sinx=[]
cosx=[]
for i in x:
    sinx.append(math.sin(i))
    cosx.append(math.cos(i))
    
# 그래프 그리기 : plt.plot([x좌표],[y좌표],'포맷 문자열')
# x 좌표를 생략하면 [0,1,2,3...]이 기본값
# x축, y축 범위 지정 : plt.axis([xmin,xmax,ymin,ymax])
plt.plot(x,sinx,label='sin(x)',color='blue')
plt.plot(x,cosx,label='cos(x)',color='red')
# 축 레이블 : plt.xlabel('x축레이블',[labelpad,loc,fontdict]) // plt.ylabel('y축레이블',[labelpad,loc,fontdict])
# labelpad : 여백 , loc : x - left/center/right, y - top/center/bottom, fontdict : 폰트
plt.xlabel('X')
plt.ylabel('Y')
# 범례 표시하기 : plt.plot([x좌표],[y좌표],label='')
# plt.legend(loc=() ncol=열의개수,fontsize,frameon,shadow)
# loc : (0,0) - 좌하단 (1,1) - 우상단 'best' - 최적위치
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
plt.legend(loc='best')
plt.title('Sine and Cosine Fuctions')
# 그리드 표시
plt.grid()
# 그래프 표시
plt.show()
```
