```python
# 4 : sin & cos  
import matplotlib.pyplot as plt # Matplotlib 
import numpy as np # numpy 
import math

# 
x = np.linspace(0,10,100)
sinx=[]
cosx=[]
for i in x:
    sinx.append(math.sin(i))
    cosx.append(math.cos(i))
    
#   : plt.plot([x],[y],' ')
# x   [0,1,2,3...] 
# x, y   : plt.axis([xmin,xmax,ymin,ymax])
plt.plot(x,sinx,label='sin(x)',color='blue')
plt.plot(x,cosx,label='cos(x)',color='red')
#   : plt.xlabel('x',[labelpad,loc,fontdict]) // plt.ylabel('y',[labelpad,loc,fontdict])
# labelpad :  , loc : x - left/center/right, y - top/center/bottom, fontdict : 
plt.xlabel('X')
plt.ylabel('Y')
#   : plt.plot([x],[y],label='')
# plt.legend(loc=() ncol=,fontsize,frameon,shadow)
# loc : (0,0) -  (1,1) -  'best' - 
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
plt.legend(loc='best')
plt.title('Sine and Cosine Fuctions')
#  
plt.grid()
#  
plt.show()
```
