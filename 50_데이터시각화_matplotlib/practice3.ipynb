```python
# 실습1 : 파이차트 그리기
import matplotlib.pyplot as plt # Matplotlib 불러오기

# 변수
labels = ['Apple', 'Banana', 'Mango', 'Blueberry']
sizes = [15, 30, 45, 10] 
explode = (0, 0.1, 0, 0)  # 중심 벗어나는 정도 : Banana 부분에서만 강조하기 위해
wedgeprops={'width':1,'edgecolor':'k','linewidth':0} # 부채꼴 스타일

# 그래프 그리기 : plt.pie([비율],lables,autopct,explode,colors,wedgeprops)
# labels : 레이블 // autopct : 소수점 자리 // colors : 색 // shadow : 그림자 // startangle = 시작 각도?
plt.pie(sizes,labels=labels,autopct='%.1f%%',explode=explode,colors=['red', 'yellow', 'green', 'purple'], wedgeprops=wedgeprops, startangle=90, shadow=True )
# 그래프 출력
plt.show()
```
