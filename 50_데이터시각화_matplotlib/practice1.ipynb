```python
# 1 :  
import matplotlib.pyplot as plt # Matplotlib 
# 
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales_2019 = [100,120,140,110,130,150,160,170,180,200,190,210]
sales_2020 = [90,110,130,120,140,160,170,160,150,180,200,190]
#   : plt.plot([x],[y],' ')
# x   [0,1,2,3...] 
# x, y   : plt.axis([xmin,xmax,ymin,ymax])
plt.plot(months,sales_2019,label='2019')
plt.plot(months,sales_2020,label='2020')
#   : plt.xlabel('x',[labelpad,loc,fontdict]) // plt.ylabel('y',[labelpad,loc,fontdict])
# labelpad :  , loc : x - left/center/right, y - top/center/bottom, fontdict : 
plt.xlabel('Month')
plt.ylabel('Sales')
#   : plt.plot([x],[y],label='')
# plt.legend(loc=() ncol=,fontsize,frameon,shadow)
# loc : (0,0) -  (1,1) -  'best' - 
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html
plt.legend(loc='best')
plt.title('Monthly Sales Comparison (2019-2020)')
#  
plt.show()
```
