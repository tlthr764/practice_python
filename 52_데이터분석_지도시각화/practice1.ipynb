```python
import folium # folium 불러오기
# 지도 생성 : m = folium.Map(location=[위도,경도],zoom_start=확대정도,tiles=타일스타일,width=가로,height=세로)
# zoom_start의 최대값은 18이다
# 타일 스타일 종류 : openstreetmap(default) // cartodbpositron // cartodbdark_matter
map = folium.Map(location=[37.547631, 126.942463], zoom_start=15)
# 마커 추가 : folium.Marker([위도,경도],popup="문구",tooltip="문구",icon=folium.Icon(color='색상',icon='아이콘')).add_to(m)
folium.Marker([37.5471,126.9425],popup='맛집',tooltip='대흥칼국수').add_to(map)
folium.Marker([37.5476,126.9428],popup='카페',tooltip='소으니').add_to(map)
folium.Marker([37.5487,126.9438],popup='학교',tooltip='서울용강초등학교').add_to(map)

map
```
